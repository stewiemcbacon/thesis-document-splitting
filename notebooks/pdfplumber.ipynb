```python
import pdfplumber
import re
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Corpus1
freqs1 = []
dir = 'corpus1/TrainTestSet/Trainset/data/'
for filename in os.listdir(dir):
    f = os.path.join(dir, filename)
    if os.path.isfile(f):
        print(f)
        with pdfplumber.open(f) as pdf:
            for i in range(len(pdf.pages)):
                freqs1.append(len(pdf.pages[i].extract_text().replace('\n','').split()))

freqs1 = pd.Series(freqs1)
```

```python
# Corpus2
freqs2 = []
dir = 'corpus2/TrainTestSet/Trainset/data/'
for filename in os.listdir(dir):
    f = os.path.join(dir, filename)
    if os.path.isfile(f):
        print(f)
        with pdfplumber.open(f) as pdf:
            for i in range(len(pdf.pages)):
                try:
                    freqs2.append(len(pdf.pages[i].extract_text().replace('\n','').split()))
                except:
                    continue

freqs2=pd.Series(freqs2)
```

```python
sns.kdeplot(freqs1[freqs1>0],log_scale=True,color='red')
sns.kdeplot(freqs2[freqs2>0],log_scale=True,color='green')
plt.legend(['corpus1','corpus2'])
plt.xlabel('Amount of words')
```
