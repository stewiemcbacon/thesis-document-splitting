```python
def compare_sim(df):
    _1 = df.shift()
    sim_scores = [0]

    for i in tqdm(range(1,len(df))):
        a = model.encode(df_1.iloc[i]['text_y_cleaned'])
        b = model.encode(_1.iloc[i]['text_y_cleaned'])
        score = util.cos_sim(a,b).item()
        sim_scores.append(score)
    return sim_scores
```

```python
df_1['text_bert_sim_score'] = compare_sim(df_1)
```
